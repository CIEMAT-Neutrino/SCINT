```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
taus = pd.read_csv("../Jan22_taus.txt", header=0, sep = "\t")
print(taus)
```

```python
SiPM0 = taus["Tauslow"][taus["Detector"]=="SiPM0"]
SiPM1 = taus["Tauslow"][taus["Detector"]=="SiPM1"]
PMT = taus["Tauslow"][taus["Detector"]=="PMT"]
counter = np.arange(0,9)
print(PMT)
PMT_1200 = [PMT[18],PMT[21],PMT[24]]
PMT_1300 = [PMT[19],PMT[22],PMT[25]]
PMT_1400 = [PMT[20],PMT[23],PMT[26]]
```

```python
# Plot de los datos y el ajuste
plt.plot(counter, SiPM0, label='SiPM#289', marker = "o")
plt.plot(counter, SiPM1, label='SiPM#290', marker = "o")
plt.plot([0,3,6], PMT_1200, label='PMT (1200V)', marker = "o")
plt.plot([0,3,6], PMT_1300, label='PMT (1300V)', marker = "o")
plt.plot([0,3,6], PMT_1400, label='PMT (1400V)', marker = "o")
plt.axvline(2.5, color = "black", linestyle = "dotted", alpha = 0.8)
plt.axvline(5.5, color = "black", linestyle = "dotted", alpha = 0.8)
# plt.errorbar(x, gain_ch1, yerr=Dgain_ch1, label='SiPM#290', marker = "o")
# plt.errorbar(x, gain_ch3, yerr=Dgain_ch3, label='SiPM#377', marker = "o")
# plt.errorbar(x, gain_ch4, yerr=Dgain_ch4, label='SiPM#378', marker = "o")

plt.xlabel("counter"); plt.ylabel("Tau slow (s)"); plt.title("January purity results")
plt.grid(True)
plt.legend()
plt.savefig("taus.png", dpi = 400)
plt.show()
```

```python
OPT  = {
    "MICRO_SEC":   True,
    "NORM":        False,                # Runs can be displayed normalised (True/False)
    "ALIGN":       True,
    "LOGY":        True,               # Runs can be displayed in logy (True/False)
    "SHOW_AVE":    "",             # If computed, vis will show average (AveWvf,AveWvfSPE,etc.)
    "SHOW_PARAM":  False,                 # Print terminal information (True/False)
    "CHARGE_KEY":  "ChargeAveRange",     # Select charge info to be displayed. Default: "ChargeAveRange" (if computed)
    "PEAK_FINDER": False,                # Finds possible peaks in the window (True/False)
    "LEGEND":      True,                # Shows plot legend (True/False)
    "SHOW":        True
    }
```

```python
Time = np.arange(len(my_runs_Jan[12][0]["AveWvf"][0]))*4e-9
Jan_Alpha_SiPM0 = my_runs_Jan[12][0]["AveWvf"][0]
VIS_Alpha_SiPM0 = my_runs_XA_VIS[6][0]["AveWvf"][0]

Jan_Laser_SiPM0 = my_runs_Jan[55][0]["AveWvf"][0]
VIS_Laser_SiPM0 = my_runs_XA_VIS[14][0]["AveWvf"][0]
```

```python
ref_max_idx = np.argmax(Jan_Alpha_SiPM0)
idx_move = np.argmax(VIS_Alpha_SiPM0)
VIS_Alpha_SiPM0 = shift(VIS_Alpha_SiPM0, ref_max_idx-idx_move, cval = 0)

VIS_Alpha_SiPM0 = VIS_Alpha_SiPM0/np.max(VIS_Alpha_SiPM0) * np.max(Jan_Alpha_SiPM0)
```

```python
ref_max_idx = np.argmax(VIS_Laser_SiPM0)
idx_move = np.argmax(Jan_Laser_SiPM0)
Jan_Laser_SiPM0 = shift(Jan_Laser_SiPM0, ref_max_idx-idx_move, cval = 0)

Jan_Laser_SiPM0 = Jan_Laser_SiPM0/np.max(Jan_Laser_SiPM0) * np.max(VIS_Laser_SiPM0)
```

```python
df = pd.DataFrame({"Time":Time,"Jan_Alpha_SiPM0":Jan_Alpha_SiPM0, "VIS_Alpha_SiPM0":VIS_Alpha_SiPM0,
                   "Jan_Laser_SiPM0":Jan_Laser_SiPM0, "VIS_Laser_SiPM0":VIS_Laser_SiPM0})
```

```python
fig = px.line(df, x="Time", y=["Jan_Laser_SiPM0", "VIS_Laser_SiPM0"], log_y = True,height=700, width = 900)
fig.show()
```
