```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python

def baricenters(input_coordinates):
    """Needs at least 4 points"""
    if (input_coordinates.shape[0]<3): 
        raise Exception("Need at least 3 points for the area")
    if (not input_coordinates.shape[1]==2): 
        raise Exception("should have 2 columns, X and Y, wrong dimensions")

    #  x_coord = input_coordinates[:,0]
    #  y_coord = input_coordinates[:,1]
    baricent=[]
    
    N_triangles=input_coordinates.shape[0]-2
    
    ref_x=input_coordinates[:,0][0]/3
    ref_y=input_coordinates[:,1][0]/3
    
    for i in range(N_triangles):        
        baricent.append([
                         np.sum(input_coordinates[:,0][i+1:i+3] )/3+ref_x,#x
                         np.sum(input_coordinates[:,1][i+1:i+3] )/3+ref_y,#y
                        ])
    return np.array(baricent)

```
