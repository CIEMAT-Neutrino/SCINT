# Pretrigger behavior

```python
run=1;
ch=0;
Lim_ped=my_runs[run][ch]['RawPedLim'];
RawPedMean_Mean=np.mean(my_runs[run][ch]['RawPedMean']);
RawPretrigMean_Mean=np.mean(my_runs[run][ch]['RawPreTriggerMean']);
PreTrigMean=np.mean(my_runs[run][ch]['RawADC'][:,:Lim_ped],axis=0);
```

```python
fig=go.Figure();
fig.add_trace(go.Scatter(x=np.arange(len(PreTrigMean)),y=PreTrigMean,name='PreTrigger Mean'))
fig.add_trace(go.Scatter(x=[np.min(np.arange(len(PreTrigMean))), np.max(np.arange(len(PreTrigMean)))],y=[RawPedMean_Mean, RawPedMean_Mean],line_color='Green',name='RawPedMean_Mean',showlegend=True))
fig.add_trace(go.Scatter(x=[np.min(np.arange(len(PreTrigMean))), np.max(np.arange(len(PreTrigMean)))],y=[RawPretrigMean_Mean, RawPretrigMean_Mean],line_color='red',name='RawPreTrigMean_Mean',showlegend=True))
fig.update_layout(template='gridon',title='PreTrigger Mean, Run='+str(run)+' ch='+str(ch),showlegend=True)
fig.update_yaxes(title='ADC counts')
fig.update_xaxes(title='Ticks')
fig.show()
```

Probamos a calcular la diferencia entre máximo y mínimo de los pretrigger y comparar para todos los eventos.

```python
run=1;
ch=0;
Diff=my_runs[run][ch]['RawPreTriggerMax']-my_runs[run][ch]['RawPreTriggerMin'];

fig1=go.Figure()
fig1.add_trace(go.Histogram(x=Diff))
fig1.show()
```

View Rare Events

```python
rare_ev=np.where(Diff>3*np.std(Diff))[0];
ev=np.random.choice(rare_ev);

fig1=go.Figure();
fig1.add_trace(go.Scatter(x=np.arange(len(my_runs[run][ch]['RawADC'][ev])),y=my_runs[run][ch]['RawADC'][ev]))
fig1.add_trace(go.Scatter(x=[np.min(np.arange(len(my_runs[run][ch]['RawADC'][ev]))),np.max(np.arange(len(my_runs[run][ch]['RawADC'][ev])))],y=[my_runs[run][ch]['RawPedMean'][ev], my_runs[run][ch]['RawPedMean'][ev]],name='RawPedMean'))
fig1.add_trace(go.Scatter(x=[np.min(np.arange(len(my_runs[run][ch]['RawADC'][ev]))),np.max(np.arange(len(my_runs[run][ch]['RawADC'][ev])))],y=[my_runs[run][ch]['RawPreTriggerMean'][ev],my_runs[run][ch]['RawPreTriggerMean'][ev]], name='RawPreTriggerMean'))
fig1.update_layout(title='PreTrigger, Run='+str(run)+', channel='+str(ch)+', event='+str(ev),title_x=0.5)
```
