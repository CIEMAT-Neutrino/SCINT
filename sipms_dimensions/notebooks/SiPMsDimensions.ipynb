```python
# Importing the function and packages to be used. Make sure you have installed them!
import sys; sys.path.insert(0, '../'); from lib import *
```

```python
# Labels for the columns
pcbs_labels = ['Diametro Taladro 1 (S1)', 'Posición Y S1', 'Posición X S1', 'Diametro Taladro 2 (S6)', 'Posición Y TS6', 'Posición X TS6', 'Planitud PCB','Anchura (media)', 'Anchura (maxima)', 'Longitud (media)','Longitud (maxima)', 'Longitud rebaje izquierdo', 'Anchura rebaje izquierdo', 'Longitud rebaje derecho', 'Anchura rebaje derecho', 'Distancia entre taladros']
sipm_labels = ['Anchura', 'Longitud', 'Posición X', 'Posición Y', 'Planitud', 'Altura']
pina_labels = ['Posición X', 'Posición Y', 'Altura Pin'] #pin anverso
pinr_labels = ['Altura Pin'] #pin reverso
```

```python
# Generating data_frames for the pcbs, sipm, pins_anverso/reverso. 
# You will need to change the folder where you are storing the data
# If save = True you will have .txts in the fir_data file with these results
# They can also be printed in the notebook with terminal_output = True

folder = "../data/10_03_2023/"
pcbs,sipm,pinr,pina = data2npy(folder=folder,pcbs_labels=pcbs_labels,sipm_labels=sipm_labels,pins_labels=pina_labels,mode=10,debug=False)

df_pcbs = df_display(pcbs,labels = pcbs_labels,name="pcbs",terminal_output=True,save=True)
df_sipm = df_display(sipm,labels = sipm_labels,name="sipm",terminal_output=True,save=True,index=["SiPM #1","SiPM #2","SiPM #3","SiPM #4","SiPM #5","SiPM #6"]*len(pcbs))
df_pina = df_display(pina,labels = pina_labels,name="pina",terminal_output=True,save=True,index=["Pin #1","Pin #2","Pin #3","Pin #4","Pin #5","Pin #6","Pin #7","Pin #8"]*len(pcbs))
df_pinr = df_display(pinr,labels = pinr_labels,name="pinr",terminal_output=True,save=True,index=["Pin #1","Pin #2","Pin #3","Pin #4","Pin #5","Pin #6","Pin #7","Pin #8"]*len(pcbs))
```

```python
# SANITY CHECK: to confirm we havent duplicated data in the arrays#
print("---- PCBs ----")
sanity_check(df_pcbs,pcbs)

print("---- SiPM ----")
sanity_check(df_sipm,sipm)

print("---- Pins_anv ----")
sanity_check(df_pina,pina)

print("---- Pins_rev ----")
sanity_check(df_pinr,pinr)
```

```python
# Generating data_frames for MEAN+STD of the previous values for the pcbs, sipm, pins_anverso/reverso. 
pcbs_mean, pcbs_std = npy2df(pcbs)
sipm_mean, sipm_std = npy2df(sipm)
pinr_mean, pinr_std = npy2df(pinr)
pina_mean, pina_std = npy2df(pina)

df_pcbs_mean = df_display(np.array((pcbs_mean,pcbs_std)),labels = pcbs_labels,index=["Mean","STD"],name="pcbs_mean",terminal_output=True,save=True)
df_sipm_mean = df_display(np.array((sipm_mean,sipm_std)),labels = sipm_labels,index=["Mean","STD"],name="sipm_mean",terminal_output=True,save=True)
df_pina_mean = df_display(np.array((pina_mean,pina_std)),labels = pina_labels,index=["Mean","STD"],name="pina_mean",terminal_output=True,save=True)
df_pinr_mean = df_display(np.array((pinr_mean,pinr_std)),labels = pinr_labels,index=["Mean","STD"],name="pinr_mean",terminal_output=True,save=True)
```

```python
# FINAL STEP: plots
# Giving titles, labels and the previous generated dfs

titles = ['Pins Height (back)','Pins Height (front)','PCB Width','SiPMs Height']
xlabel = ['Height [mm]','Height [mm]','Width [mm]','Height [mm]']
ylabel = ['Nº Pins','Nº Pins','Nº PCB','Nº SiPMs']
df_raw = [df_pinr,df_pina,df_pcbs,df_sipm]
df_fin = [df_pinr_mean,df_pina_mean,df_pcbs_mean,df_sipm_mean]
colums = ['Altura Pin','Altura Pin','Anchura (media)','Altura']
colors = [['C0','purple','orange'],['C0','purple','orange'],['C0','purple','orange'],['C0','purple','orange']]
for i in range(len(titles)):
    plotitos(titles[i],xlabel[i],ylabel[i],df_raw[i],df_fin[i],colums[i],colors[i])
```
