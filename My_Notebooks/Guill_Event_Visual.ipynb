```python
INPUT_FILE = "MegaCell_LAr"; OV = 0 #To Select the run: OV=0 -> run=103; OV=1 -> run=113; OV=2 -> run=105; OV=3 -> run=111  
PRESET ="ALL"
info = read_input_file(INPUT_FILE)  # Read input file
channels = [0,1,6,7]

#-------------------------------- LOAD RUNS ---------------------------------#
run_keys = ["CALIB_RUNS","ALPHA_RUNS","MUONS_RUNS","NOISE_RUNS"]
nruns = dict.fromkeys(run_keys)
for key in run_keys:
    try:               nruns[key] = info[key][OV] # Store runs in dictionary
    except IndexError: nruns.pop(key)
print(nruns)

runs = dict.fromkeys(nruns.keys())
for run in runs: runs[run] = load_npy(np.asarray([nruns[run]]).astype(int),np.asarray(channels).astype(int),preset=PRESET,info=info,compressed=True)
#----------------------------------------------------------------------------#
```

```python
RUN2PLOT = "CALIB_RUNS";
ADC_mean={};
Lim_ped={};
for c in channels:
    ADC_mean[c]=np.mean(runs[RUN2PLOT][runs[RUN2PLOT]["NRun"][0]][c]["RawADC"],axis=0);
    Lim_ped[c]=runs[RUN2PLOT][runs[RUN2PLOT]["NRun"][0]][c]["RawPedLim"];
```

```python
ch=np.random.choice(channels);
lim_pedes=Lim_ped[ch]*runs[RUN2PLOT][runs[RUN2PLOT]["NRun"][0]][ch]["Sampling"];
print(lim_pedes)
```

```python
ch=np.random.choice(channels);
event=np.random.randint(len(runs[RUN2PLOT][runs[RUN2PLOT]["NRun"][0]][ch]["RawADC"]));
ADC=runs[RUN2PLOT][runs[RUN2PLOT]["NRun"][0]][ch]["RawADC"][event]
time=np.arange(0,len(runs[RUN2PLOT][runs[RUN2PLOT]["NRun"][0]][ch]["RawADC"][event])*runs[RUN2PLOT][runs[RUN2PLOT]["NRun"][0]][c]["Sampling"],runs[RUN2PLOT][runs[RUN2PLOT]["NRun"][0]][c]["Sampling"])
lim_pedes=Lim_ped[ch]*runs[RUN2PLOT][runs[RUN2PLOT]["NRun"][0]][ch]["Sampling"]

fig=go.Figure()
fig.add_trace(go.Scatter(x=time,y=ADC,name="Channel "+str(ch)+" Event="+str(event)))
fig.add_trace(go.Scatter(x=time,y=ADC_mean[ch],name="Channel "+str(ch)+" mean"))
fig.add_vline(x=lim_pedes,name='Ped Limit'+' Channel '+str(ch))
fig.update_layout(title='Event '+str(event)+' channel '+str(ch)+' vs Mean',title_x=0.5, xaxis_title='Time [s]', yaxis_title='ADC counts')
fig.show()
```
