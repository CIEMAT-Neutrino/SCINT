````python
# Define a single Gaussian function
def gaussian(x, amplitude, mean, std_dev):
    return amplitude * np.exp(-(x - mean)**2 / (2 * std_dev**2))

# Define a function to model a train of Gaussian peaks
def multiple_gaussians(x, *params):
    num_peaks = len(params) // 3  # Each peak has three parameters: amplitude, mean, and standard deviation
    result = np.zeros_like(x)
    
    for i in range(num_peaks):
        amplitude, mean, std_dev = params[i * 3:i * 3 + 3]
        result += gaussian(x, amplitude, mean, std_dev)
    
    return result

# Generate some example data
x_data = np.linspace(0, 10, 200)
true_params = [1.0, 2.0, 0.5, 0.7, 4.0, 0.3]  # Amplitude, mean, and std_dev for two Gaussian peaks
y_data = multiple_gaussians(x_data, *true_params) + 0.1 * np.random.randn(len(x_data))

# Initial guess for the parameters
initial_guess = [1.0, 1.5, 0.2, 0.5, 3.5, 0.2]  # Amplitude, mean, and std_dev for two Gaussian peaks

# Fit the data to the model
fit_params, _ = curve_fit(multiple_gaussians, x_data, y_data, p0=initial_guess, maxfev=5000)

# Plot the original data and the fitted curve
plt.figure(figsize=(8, 6))
plt.plot(x_data, y_data, 'b', label='Data')
plt.plot(x_data, multiple_gaussians(x_data, *fit_params), 'r', label='Fitted Curve')
plt.legend(loc='best')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Fitting Multiple Gaussians')
plt.show()

# Display the fitted parameters
print("Fitted Parameters:")
for i in range(len(fit_params) // 3):
    amplitude, mean, std_dev = fit_params[i * 3:i * 3 + 3]
    print(f"Peak {i + 1}: Amplitude = {amplitude:.3f}, Mean = {mean:.3f}, Std Dev = {std_dev:.3f}")
````

````python
def fit_gaussians(x, y, *p0):
    assert x.shape == y.shape, "Input arrays must have the same shape."
    # try:
    popt, pcov = curve_fit(gaussian_train, x,y, p0=p0[0], maxfev=8000)
    fit_y=gaussian_train(x,*popt)
    chi_squared = np.sum((y[abs(fit_y)>0.1] - fit_y[abs(fit_y)>0.1]) ** 2 / fit_y[abs(fit_y)>0.1]) / (y.size - len(popt))
    return popt,fit_y,chi_squared
````
