```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Datos de ejemplo
x = np.array([3,4,5])
```

```python
gains = pd.read_csv("../gain_XA_VUV.txt", header=0, sep = "\t")
print(gains)
```

```python
gain_ch0 = gains["Gain"][gains["Ch"]==0]
Dgain_ch0 = gains["DGain"][gains["Ch"]==0]

gain_ch1 = gains["Gain"][gains["Ch"]==1]
Dgain_ch1 = gains["DGain"][gains["Ch"]==1]

gain_ch3 = gains["Gain"][gains["Ch"]==3]
Dgain_ch3 = gains["DGain"][gains["Ch"]==3]

gain_ch4 = gains["Gain"][gains["Ch"]==4]
Dgain_ch4 = gains["DGain"][gains["Ch"]==4]

gain_ch5 = gains["Gain"][gains["Ch"]==5]
Dgain_ch5 = gains["DGain"][gains["Ch"]==5]
```

```python
# Ajuste lineal
a0, b0 = np.polyfit(x, gain_ch0, 1)
a1, b1 = np.polyfit(x, gain_ch1, 1)
a3, b3 = np.polyfit(x, gain_ch3, 1)
a4, b4 = np.polyfit(x, gain_ch4, 1)

# Plot de los datos y el ajuste
plt.errorbar(x, gain_ch0, yerr=Dgain_ch0, label='SiPM#289', marker = "o")
plt.errorbar(x, gain_ch1, yerr=Dgain_ch1, label='SiPM#290', marker = "o")
plt.errorbar(x, gain_ch3, yerr=Dgain_ch3, label='SiPM#377', marker = "o")
plt.errorbar(x, gain_ch4, yerr=Dgain_ch4, label='SiPM#378', marker = "o")

plt.xlabel("OV (V)"); plt.ylabel("Gain"); plt.title("SiPMs gain")
plt.grid(True)
plt.legend()
plt.savefig("gain_fit_SiPMs.png", dpi = 400)
plt.show()
```

```python
# Ajuste lineal
a5, b5 = np.polyfit(x, gain_ch5, 1)

# Plot de los datos y el ajuste
plt.errorbar(x, gain_ch5, yerr=Dgain_ch5, label='SiPM#289', marker = "o")

plt.xlabel("OV (V)"); plt.ylabel("Gain"); plt.title("APSAIA_XA_VUV gain")
plt.grid(True)
plt.legend()
plt.savefig("gain_fit_XA_VUV.png", dpi = 400)
plt.show()
```
