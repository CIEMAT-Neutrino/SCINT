## Multiple Runs plots

### STD

```python
%matplotlib inline

# WEEK="APSAIA_VIS"   # 1st  week 
# WEEK="APSAIA_VUV"   # 2nd  week 
WEEK="DAPHNE_VIS"   # 3rd  week 
# WEEK="DAPHNE_VUV"   # 4th  week 
# WEEK="APSAIA_VUV_2" # 5th  week 
path="/scr/neutrinos/rodrigoa/"+WEEK+"/joython/"
Runs=open_runs_table("../macros/"+WEEK+".xlsx")


# path="/media/rodrigoa/DiscoDuro/SBND_XA_PDE/APSAIA_VUV/joython/"


mean=np.zeros((len(Runs["Channels"].iloc[0]),Runs["Run"].array.shape[0]))
ress=np.zeros((len(Runs["Channels"].iloc[0]),Runs["Run"].array.shape[0]))

i=0
for ch in Runs["Channels"].iloc[0]:
    r=0;
    for run in Runs["Run"].array:
        if r==3:        r+=1; continue
        
        Run_props=Runs[Runs["Run"]==run].iloc[0]
        run_path=path+"run"+str(run).zfill(2)+"/";
        compress=False
        Pedestal_vars=open_run_var(run_path,"Pedestal_vars_SW",[ch],compressed=compress)
        
        counts, bins=np.histogram(Pedestal_vars[ch]["STD"], 300,[0,20])
        mode=bins[np.argmax(counts)]
        resolution=np.std(Pedestal_vars[ch]["STD"][Pedestal_vars[ch]["STD"]<50])
        mean[i,r]=mode
        ress[i,r]=resolution
        r+=1;
    i+=1;


x=Runs["Run"].array
ys=mean
yer=ress
legend=Run_props["ChannelName"]

fig,axs=horizontal_subplots(x,ys,yer,legend)
axs[0].set_ylim(0,30)
axs[0].set_xlabel(  "Pedestal STD [ADCs]"  )
axs[0].set_title(  WEEK  )
# noise_runs=(Runs["Type"]=="Noise").array.to_numpy()
# axs[-1].plot(x[noise_runs].to_numpy(),ys[-1][noise_runs],'x',color="tab:orange")
```

### Mean

```python
%matplotlib inline

for WEEK in ["APSAIA_VIS","APSAIA_VUV","DAPHNE_VIS","DAPHNE_VUV","APSAIA_VUV_2"]:
# WEEK="APSAIA_VIS"   # 1st  week 
# WEEK="APSAIA_VUV"   # 2nd  week 
# WEEK="DAPHNE_VIS"   # 3rd  week 
# WEEK="DAPHNE_VUV"   # 4th  week 
# WEEK="APSAIA_VUV_2" # 5th  week 
    path="/scr/neutrinos/rodrigoa/"+WEEK+"/joython/"
    Runs=open_runs_table("../macros/"+WEEK+".xlsx")


    # path="/media/rodrigoa/DiscoDuro/SBND_XA_PDE/APSAIA_VUV/joython/"


    mean=np.zeros((len(Runs["Channels"].iloc[0]),Runs["Run"].array.shape[0]))
    ress=np.zeros((len(Runs["Channels"].iloc[0]),Runs["Run"].array.shape[0]))

    i=0
    for ch in Runs["Channels"].iloc[0]:
        r=0;
        for run in Runs["Run"].array:
            
            Run_props=Runs[Runs["Run"]==run].iloc[0]
            run_path=path+"run"+str(run).zfill(2)+"/";
            compress=False
            Pedestal_vars=open_run_var(run_path,"Pedestal_vars_SW",[ch],compressed=compress)
            
            counts, bins=np.histogram(Pedestal_vars[ch]["MEAN"], 1000)
            mode=bins[np.argmax(counts)]
            resolution=np.std(Pedestal_vars[ch]["MEAN"][Pedestal_vars[ch]["STD"]<50])
            mean[i,r]=mode
            ress[i,r]=resolution
            r+=1;
        i+=1;

    offset=1.26e4
    mean[:-1]-=offset
    x=Runs["Run"].array
    ys=mean
    yer=ress
    legend=Run_props["ChannelName"]

    fig,axs=horizontal_subplots(x,ys,yer,legend)
    # axs[0].set_ylim(1.545e4,1.565e4)
    axs[0].set_xlabel(  "Pedestal Mean [ADCs],\n SiPMs PedMean =- {:.0f}".format(offset)  )
    axs[0].set_title(  WEEK) 
    plt.suptitle("Baseline stability over runs", fontsize=14)
    # noise_runs=(Runs["Type"]=="Noise").array.to_numpy()
    # axs[-1].plot(x[noise_runs].to_numpy(),ys[-1][noise_runs],'x',color="tab:orange")
    # plt.yscale("log")
```
